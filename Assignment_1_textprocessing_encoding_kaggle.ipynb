# Remove stopwords

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("all")
```

```python
def remove_stopwords(text):
    new_text=[]

    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())


    return " ".join(new_text).replace("   ","")
```

```python
dataset["text_processed"]=dataset["text_processed"].apply(remove_stopwords)
```

# Remove punctuation

```python
import string
```

```python
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
dataset["text_processed"]=dataset["text_processed"].apply(remove_punc)
```

# Stemming

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word

dataset["stemming"] = dataset["text_processed"].apply(stemming)
```

```python
dataset.filter(["text_processed","stemming"])
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer

def lemmatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=[lemmetizer.lemmatize(word) for word in words]

    return lemetized_word
```

```python
dataset["lemmatising"] = dataset["text_processed"].apply(lemmatization)
dataset.filter(["text_processed","stemming","lemmatising"])
```

# 5 Advantages and Disadvantage of OHE Vs BOW Vs Ngram Vs TF-IDF


**One-Hot Encoding (OHE):**

*Advantages*:


1.   Simple and Intuitive: OHE is a straightforward technique where each word is represented as a binary feature.
2.   Independence: OHE treats each word as an independent entity, which might be suitable for certain tasks.

*Disadvantages:*


1.   High Dimensionality: OHE results in a high-dimensional and sparse representation, which can be computationally expensive and memory-intensive.
2.   Loss of Word Order: It does not consider the order of words in a document, leading to a loss of sequential information.
3.   Context is lost
4.   Rare word and common word are captured equally there is not differentiation as such





**Bag-of-Words (BOW):**

*Advantages:*


1.   Simplicity: BOW is easy to implement and understand.
2.   Versatility: It is effective in various text classification tasks.
3.   Insensitivity to Word Order: BOW is insensitive to the order of words, making it suitable for tasks where word order is not crucial.

*Disadvantages:*


1.   Loss of Sequence Information: Like OHE, BOW loses information about the sequence of words in a document.
2.   Frequency Emphasis: It may emphasize frequent words that may not be informative.






**N-grams:**

*Advantages:*


1.   Capture Local Context: N-grams capture local context by considering adjacent words.
2.   Retain Some Order Information: Unlike BOW, N-grams retain some order information.


*Disadvantages:*


1.   High Dimensionality: As the size of N increases, the dimensionality of the feature space grows exponentially.
2.   Sparse Representation: Like OHE and BOW, N-grams can result in a sparse representation.





**Term Frequency-Inverse Document Frequency (TF-IDF):**

*Advantages:*

1. Term Importance: TF-IDF considers the importance of terms by assigning higher weights to less frequent but discriminative words.
2. Effective in Information Retrieval: It is widely used in information retrieval tasks.

*Disadvantages:*

1. Ignores Word Order: Similar to BOW and OHE, TF-IDF ignores the order of words in a document.
2. Sensitivity to Tokenization: The choice of tokenization can affect TF-IDF results.
